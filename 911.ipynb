## Análise do Horário e Data:

```python
print(type(df_911['timeStamp'].iloc[0]))
df_911.info()
```

```python
df_911['timeStamp'] = pd.to_datetime(df_911['timeStamp'])
df_911.info()
```

```python
print("Ocorrência mais Recente: ",df_911['timeStamp'].max())
print("Ocorrência mais Antiga: ",df_911['timeStamp'].min())
```

### Separando:

```python
# Horas e Minutos:
df_911['hour'] = df_911['timeStamp'].apply(lambda hora: hora.hour)
df_911['minute'] = df_911['timeStamp'].apply(lambda minuto: minuto.minute)
# Dia, Mês e Ano
df_911['day'] = df_911['timeStamp'].apply(lambda dia: dia.day)
df_911['dayofweek'] = df_911['timeStamp'].apply(lambda dia: dia.dayofweek) # ===> transforma o dia para o dia referente da semana, partindo do 0 (Segunda) a 6(Domingo).
df_911['month'] = df_911['timeStamp'].apply(lambda mes: mes.month)
df_911['year'] = df_911['timeStamp'].apply(lambda ano: ano.year)
```

### Mapeando o DataSet:

```python
dia_mapeado = {0:'Seg',
               1:'Ter',
               2: 'Qua',
               3: 'Qui',
               4: 'Sex',
               5: 'Sab',
               6: 'Dom'}
```

```python
df_911['DiaSemana'] = df_911['dayofweek'].map(dia_mapeado)
```

```python
ordem = ['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'Sab', 'Dom']

plt.figure(figsize=(10,6))
sns.countplot(x='DiaSemana', data=df_911, hue='Reason', order=ordem, palette='mako')
plt.title('Frequência de Tipo de Chamada por Dias da Semana')
plt.xlabel('Dia da Semana')
plt.ylabel('Nº de Chamadas')
plt.legend(bbox_to_anchor=(1,1), loc=2)
```

## Análise Agrupada:

```python
df_Fire = df_911[df_911['Reason'] == 'Fire']
df_EMS = df_911[df_911['Reason'] == 'EMS']
df_Traffic = df_911[df_911['Reason'] == 'Traffic']
```

### Acidentes com Fogo:

```python
dia_hora = df_Fire.groupby(by=['dayofweek' , 'hour']).count()['Reason'].unstack() # ===> unstack serve para empilhar os dados (no caso, dia e hora).
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(dia_hora, cmap='viridis')
plt.xlabel('Hora')
plt.ylabel('Dia da Semana')
plt.title('Acidentes com Fogo por Horas em Dias da Semana')
```

```python
plt.figure(figsize=(10,6))
sns.clustermap(dia_hora, cmap='viridis')
```

### Acidentes de Saúde:

```python
dia_hora = df_EMS.groupby(by=['dayofweek' , 'hour']).count()['Reason'].unstack() # ===> unstack serve para empilhar os dados (no caso, dia e hora).
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(dia_hora, cmap='viridis')
plt.xlabel('Hora')
plt.ylabel('Dia da Semana')
plt.title('Acidentes de Saúde por Horas em Dias da Semana')
```

```python
plt.figure(figsize=(10,6))
sns.clustermap(dia_hora, cmap='viridis')
```

### Acidentes com Trânsito:

```python
dia_hora = df_Traffic.groupby(by=['dayofweek' , 'hour']).count()['Reason'].unstack() # ===> unstack serve para empilhar os dados (no caso, dia e hora).
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(dia_hora, cmap='viridis')
plt.xlabel('Hora')
plt.ylabel('Dia da Semana')
plt.title('Acidentes de Trânsito por Horas em Dias da Semana')
```

```python
plt.figure(figsize=(10,6))
sns.clustermap(dia_hora, cmap='viridis')
```
